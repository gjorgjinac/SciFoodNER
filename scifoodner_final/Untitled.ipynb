```python
import pandas as pd
```

```python
hansard=pd.read_csv('ontology_mappings/hansard.csv',sep='|')
hansard['pure tag']=hansard['Semantic Tags'].apply(lambda x: ';'.join([y.split('[')[0].strip() for y in x.split(';')] ) )
hansard.to_csv('ontology_mappings/hansard_pure.csv')
```

```python
prefix = 'http://purl.obolibrary.org/obo/'
foodon =pd.read_csv('ontology_mappings/foodon.csv',sep='|')
foodon['URLs']=foodon['URLs'].apply(lambda x: x.replace(prefix,''))
```

```python
foodon.to_csv('ontology_mappings/foodon_pure.csv')
```

```python
foodon.drop_duplicates('URLs')
```
